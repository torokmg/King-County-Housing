# Checking How Location Affects Price

I needed to figure out a way to use zipcode data in my model. I originally thought the best way to do this would be to see how much homes sell for in each zipcode. After looking at the data though it was clear that each zipcode would have its own range of prices so it would be bias to compare them that way.  My solution was to calculate the price per sqft for each house and compare them that way.

```python
df['price_per_sqft'] = df['price']/df['sqft_living'] # adding price per sqft
```

I then grouped each house by zipcode.  After grouping I was able to calculate the mean price per sqft for each zipcode using my price_per_sqft number.

```python
df = df.reset_index() # bring back the id column so we can access it
frequency = df.groupby('zipcode').count()['id'] # group zipcodes, and sum up number of sales
mean = df.groupby('zipcode').mean()['price_per_sqft'] # get mean price/sqft for sales
cluster = pd.concat([frequency, mean], axis=1) # combine these
cluster['zipcode'] = cluster.index
cluster.columns = ['frequency', 'price_per_sqft', 'zipcode']
cluster.describe()
```

Next I split the zipcodes into three distict groups:
- Zipcodes that have sold below the average price (cheaper areas)
- Zipcodes that sell above the average price, but not often (don't want to have their values overrepresented)
- Zipcodes that sell above the average price, and more frequently.

# creating "cheap" group that sell below mean price/sqft (250)
cluster_cheap = cluster[cluster.price_per_sqft < 250]

# temporary cluster of the houses that sell above mean price/sqft
cluster_temp = cluster[cluster.price_per_sqft >= 250]

# creating "low volume, higher priced" group that sell less frequently than the mean
cluster_expensive_low_volume = cluster_temp[cluster_temp.frequency < 277]

# creating "hhigh volume, higher priced" group that sell more frequently than the mean
cluster_expensive_high_volume = cluster_temp[cluster_temp.frequency >= 277]

def get_zipcode_category(zipcode):
    if zipcode in cluster_cheap.index:
        return 'cheap'
    elif zipcode in cluster_expensive_low_volume.index:
        return 'expensive_low_volume'
    else:
        return 'expensive_high_volume'

df['zipcode_type'] = df.zipcode.apply(get_zipcode_category)

df.zipcode_type.value_counts()

I dropped the zipcode, lat, and long features as they should all be encapsulated in my new zipcode_type feature.  I will also drop the 'id' column as it is no longer needed.

```python
df = df.drop(['lat', 'long', 'zipcode', 'grade'], axis=1) # encapsulated in zipcode type
# drop id, date
```

# Creating Dummies for Categorical Data

I first wanted to bin a couple features.  The first feature was sqft_basement.  Since there were alot of 0's for this feature, I felt it would be better for the model if instead binned all 0 numbers in one column and all numbers greater than 0 would be in another.  This essentially makes the catagory a yes/no question in regards to whether a house has a basement.  The next feature I will bin is yr_renovated.  It also has alot of 0 values so I decided to do the same as I did with sqft_basement and essentially made it a yes/no question. 

Next I will create dummies for each of our catagorical features:

- zipcode_type
- view 
- condition 
- bedrooms 
- floors
- bathrooms    
- waterfront
- grade 

```python
#binning
bins = [-1 , 1 , df.sqft_basement.max()]
labels = ['0','1']

df.sqft_basement = pd.cut(df.sqft_basement, bins, labels=labels)
df.sqft_basement = df.sqft_basement.astype('category')

#make dummies
sqft_basement = pd.get_dummies(df.sqft_basement, prefix='basement')
df = df.join([sqft_basement])
df.drop(['sqft_basement'], axis=1, inplace=True)
```

```python
#binning
bins = [ -1 , 1 , df.yr_renovated.max() ]
labels = ['0','1']

df.yr_renovated = pd.cut(df.yr_renovated, bins, labels=labels)
df.yr_renovated = df.yr_renovated.astype('category')
#make dummies
yr_renovated = pd.get_dummies(df.yr_renovated, prefix='renovated')
df = df.join([yr_renovated])
df.drop(['yr_renovated'], axis=1, inplace=True)
```

```python
#binning
bins = [1899,1919,1939,1959,1979,1999,2019]

df.yr_built = pd.cut(df.yr_built, bins)
df.yr_built = df.yr_built.astype('category')
#make dummies
yr_built = pd.get_dummies(df.yr_built, prefix='built')
df = df.join([yr_built])
df.drop(['yr_built'], axis=1, inplace=True)
```

```python
#Categorize
#df.zipcode = df.zipcode.astype('category')
df.labellocation = df.labellocation.astype('category')
df.view = df.view.astype('category')
df.condition = df.condition.astype('category')
df.bedrooms = df.bedrooms.astype('category')
df.floors = df.floors.astype('category')
df.bathrooms = df.bathrooms.astype('category')
df.waterfront = df.waterfront.astype('category')
#df.grade = df.grade.astype('category')


#Make Dummies

#zipcode = pd.get_dummies(df.zipcode, prefix ='zipcode')
labellocation = pd.get_dummies(df.labellocation, prefix ='labellocation')
view = pd.get_dummies(df.view, prefix ='view')
condition = pd.get_dummies(df.condition, prefix='condition')
bedrooms = pd.get_dummies(df.bedrooms, prefix='bedrooms')
floors = pd.get_dummies(df.floors, prefix='floors')
bathrooms = pd.get_dummies(df.bathrooms, prefix='bathrooms')      
waterfront = pd.get_dummies(df.waterfront, prefix='waterfront')
#grade = pd.get_dummies(df.grade, prefix='grade')


#Add dummies remove original columns

df = df.join([waterfront, condition, bedrooms, floors, bathrooms, view, labellocation])
df.drop(['waterfront','condition','bedrooms', 'floors', 'bathrooms', 'labellocation', 'view'], axis=1, inplace=True)
```

# Modeling

Now that all of our data has been normalized and scaled, we can begin to model it.

## Model 1

```python
linreg = LinearRegression()

# our feature variables
X = df.drop(['price_per_sqft', 'price'], axis=1)

# our target variable
y = df['price']

# separate our data into testing and training subsets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)

# fitting our training feature variables to our training target variable (price)
linreg.fit(X_train, y_train)

# predicted prices from our training subset
y_hat_train = linreg.predict(X_train)

# predicted prices from our testing subset
y_hat_test = linreg.predict(X_test)

# calculating residuals
train_residuals = y_hat_train - y_train
test_residuals = y_hat_test - y_test

# calculate mean square error for our test and training results
train_mse = mean_squared_error(y_train, y_hat_train)
test_mse = mean_squared_error(y_test, y_hat_test)
print('Train Mean Squared Error:', round(train_mse,5))
print('Test Mean Squared Error:', round(test_mse,5))

# calculate r-squared
r2 = round(linreg.score(X,y), 5)
print("R Squared:", r2)

# calculate mean absolute error
mean_abs_err = metrics.mean_absolute_error(y_test, y_hat_test)
print("Mean Absolute Error:", round(mean_abs_err,5))

# calculate root mean squared error
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_hat_test))
print("Root Mean Squared Error:", round(rmse,5))

# calculate mean predicted price and mean actual price
average_predicted_price = y_hat_test.mean()
average_actual_price = y_test.mean()
print("Average Predicted Price:", round(average_predicted_price,5))
print("Average Actual Price:", round(average_actual_price,5))
print("Difference:", round(average_predicted_price - average_actual_price, 5))
```

## Model 2

For my second model, I will use feature selction to create a list of all of my features that have a p-value of under 0.05.  I will then run that list of features through my model to be able to compare it to my first model.

```python
def feature_selection(X, y):
    features = sm.OLS(y, X).fit()
    pvalues = pd.DataFrame(features.pvalues, columns=['p'])
    featureslist = list(pvalues[pvalues.p < 0.05].index)
    return featureslist
```

```python
SigFeats = feature_selection(X, y)
```

```python
df[SigFeats].info()
```

Compared to the features I used in my first model, this list is shorter.  My original model used 64 features. My second model will be using 53.  So that means that I was able to eliminate 11 features in my model that had P-values of 0.05 or greater.

```python
X_new = df[SigFeats]
y = df['price']

# separate our data into testing and training subsets
X_train, X_test, y_train, y_test = train_test_split(X_new,y, test_size=0.2, random_state=2)

# fitting our training feature variables to our training target variable (price)
linreg.fit(X_train, y_train)

# predicted prices from our training subset
y_hat_train = linreg.predict(X_train)

# predicted prices from our testing subset
y_hat_test = linreg.predict(X_test)

# calculating residuals
train_residuals = y_hat_train - y_train
test_residuals = y_hat_test - y_test

# calculate mean square error for our test and training results
train_mse = mean_squared_error(y_train, y_hat_train)
test_mse = mean_squared_error(y_test, y_hat_test)
print('Train Mean Squared Error:', round(train_mse,5))
print('Test Mean Squared Error:', round(test_mse,5))

# calculate r-squared
r2 = round(linreg.score(X_new,y), 5)
print("R Squared:", r2)

# calculate mean absolute error
mean_abs_err = metrics.mean_absolute_error(y_test, y_hat_test)
print("Mean Absolute Error:", round(mean_abs_err,5))

# calculate root mean squared error
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_hat_test))
print("Root Mean Squared Error:", round(rmse,5))

# calculate mean predicted price and mean actual price
average_predicted_price = y_hat_test.mean()
average_actual_price = y_test.mean()
print("Average Predicted Price:", round(average_predicted_price,5))
print("Average Actual Price:", round(average_actual_price,5))
print("Difference:", round(average_predicted_price - average_actual_price, 5))
```
